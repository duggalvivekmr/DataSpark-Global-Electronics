# 📊 EDA Summary – `sales_full` Dataset

## 📁 Dataset Overview
This project analyzes transactional data from a global electronics retailer. The dataset includes:

- Customer demographics
- Sales records
- Product details
- Store metrics
- Currency exchange rates

### 🛠 Engineered Features
- `Order Value` = Quantity × Unit Price USD  
- `Profit` = (Unit Price USD - Unit Cost USD) × Quantity  
- `Store Age` = Current Year - Open Year  

---

## 🔍 Customer Analysis
- **Gender**: Balanced distribution of male and female customers  
- **Age**: Majority aged between 30–60  
- **Location**: Predominantly Australian, with strong regional variation  

---

## 💰 Sales Analysis
- **Trends**: Stable monthly sales with seasonal peaks  
- **Top Products**: Desktop PCs dominate both quantity and revenue  
- **Currencies**: USD is most common; AUD and CAD are also used  

---

## 📦 Product Analysis
- **Most Popular**: Products in 'Computers' and 'Audio' categories  
- **Least Popular**: Spread across niche subcategories  
- **Profitability**: High overlap with best-selling items  
- **Category Performance**: 'Computers' lead in revenue  

---

## 🏬 Store Analysis
- **Size vs Sales**: No strong correlation; medium-sized stores perform well  
- **Top States**: A few states contribute a significant portion of sales  
- **Store Age**: Older stores show consistency but not necessarily higher revenue  

---

## 📉 Distribution & Outliers
- **Order Value**, **Unit Price**, and **Unit Cost** are right-skewed  
- **Quantity** shows left skew (most purchases are small)  
- **Outliers** visualized using histograms and boxplots — no removal needed  

---

## ✅ Summary

- Dataset is clean and enriched with engineered features  
- Sales are concentrated in a few high-performing products and regions  
- Distributions and outliers have been explored and understood
